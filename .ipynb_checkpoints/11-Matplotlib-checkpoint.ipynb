# matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny, color="red")
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny, color="red",linestyle="--")
plt.plot(x,cosy)
plt.show()
```

```python
plt.plot(x,siny, color="red",linestyle="--")
plt.plot(x,cosy)
plt.xlim(-5,15)
plt.ylim(0,1.5)
plt.show()
```

```python
plt.plot(x,siny, color="red",linestyle="--")
plt.plot(x,cosy)
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x,siny, color="red",linestyle="--")
plt.plot(x,cosy)
plt.xlabel("x ")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x,siny, color="red",linestyle="--",label="sin(x)")
plt.plot(x,cosy,label="cos(x)")
plt.xlabel("x ")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome")
plt.show()
```

## Scatter Plot

```python
plt.scatter(x,siny)
plt.show()
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color="red")
plt.show()
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.scatter(x,y,alpha=0.1)
plt.show()
```
